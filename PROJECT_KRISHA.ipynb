# **HANDLING OUTLIERS**

Outliers are data points that are significantly different from the rest of the data  unusually high or low values.

image.png

```python
num_columns=df.select_dtypes(include='number')
num_columns.head()
```

```python
for columns in num_columns.columns:
  sns.boxplot(data=num_columns,x=columns)
  plt.show()
```

```python
def handling_outliers(df):
  for col in df.select_dtypes(include='number').columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    iqr=Q3-Q1
    lower_bound=Q1-(1.5*iqr)
    upper_bound=Q3+(1.5*iqr)
    df[col]=df[col].apply(lambda x:lower_bound if x<lower_bound else upper_bound if x>upper_bound else x)
  return df
df=handling_outliers(df)
df.head()
sns.boxplot(df)
```

# **FEATURE SELECTION**

Feature selection is the process of choosing the most relevant and important variables (features) in your dataset that contribute most to the prediction of the target variable.

image.png

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
x=df1.drop(columns=["Exited"],axis=1)
y=df1["Exited"].astype(int)
```

SelectKBest is a method from sklearn that selects the top K features based on statistical scores like correlation, chi-square, or ANOVA F-test.

It ranks all features according to their relevance to the target and picks the best K features to use in your model.



```python
selecter=SelectKBest(score_func=f_classif,k='all')
x_new=selecter.fit_transform(x,y)
selecter_features=x.columns[selecter.get_support()]
feature_score=pd.DataFrame({'Feature':x.columns,'Score':selecter.scores_})
print(feature_score)
print(selecter_features)
```

```python
x_selected=x[selecter_features]
x_train,x_test,y_train,y_test=train_test_split(x_selected,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
```

# **BUILDING MODELS**

image.png

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python

Classifier={
    "Random Forest Classifier":RandomForestClassifier(),
    "Logistic Regression":LogisticRegression(),
    "Support Vector Machine":SVC(),
    "Decision Tree":DecisionTreeClassifier()
}
result={}
for name,clf in Classifier.items():
    clf.fit(x_train_scaled,y_train)
    y_pred=clf.predict(x_test_scaled)
    print(f"{name}...\n")
    print("\n training models")
    print("accuracy:",accuracy_score(y_test,y_pred))
    print("precision:",precision_score(y_test,y_pred))
    print("confusion matrix:",confusion_matrix(y_test,y_pred))
```
